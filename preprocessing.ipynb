```python
data = data.fillna('JFL')
print(data.isnull().sum())
```

```python
data['PPG'] = data['puntos_totales'] / data['partidos_jugados']
data['PPG'] = data['PPG'].round(2)
columns_to_remove = ['temporada', '5i', 'mates', 'victorias', 'derrotas', 'partidos_jugados','puntos_totales','minutos_jugados', 'puntos_maximos']
data = data.drop(columns=columns_to_remove)
print(data.columns)
```

```python
column_name = 'posicion'
replacement_dict = {
    'Base': 'B',
    'Escolta': 'E',
    'Alero': 'A',
    'Ala-pívot': 'AP',
    'Pívot': 'P'
}
data[column_name] = data[column_name].replace(replacement_dict)
data
```

```python
new_column_names = ['name', 'pos', 'height', 'birth_date', 'license', 'club', 'X3PTM', 'X3PTA', 'X3PT_Per', 'X2PTM', 'X2PTA', 'X2PT_Per', 'FTM', 'FTA', 'FT_Per', 'OR', 'DR', 'TR', 'AST', 'STL', 'TOV', 'BLKF', 'BLKA', 'FC', 'FD', 'PlusMinus', 'PIR', 'PPG']
data.columns = new_column_names
new_order = ['name', 'pos', 'height', 'birth_date', 'license', 'club', 'PPG', 'X3PTM', 'X3PTA', 'X3PT_Per', 'X2PTM', 'X2PTA', 'X2PT_Per', 'FTM', 'FTA', 'FT_Per', 'OR', 'DR', 'TR', 'AST', 'STL', 'TOV', 'BLKF', 'BLKA', 'FC', 'FD', 'PlusMinus', 'PIR']
data = data[new_order]
data
```

```python
import datetime
data['age'] = 2020 - pd.to_datetime(data['birth_date']).dt.year
data = data.drop(columns=['birth_date'])
new_order = ['name', 'pos', 'height', 'age', 'license', 'club', 'PPG', 'X3PTM', 'X3PTA', 'X3PT_Per', 'X2PTM', 'X2PTA', 'X2PT_Per', 'FTM', 'FTA', 'FT_Per', 'OR', 'DR', 'TR', 'AST', 'STL', 'TOV', 'BLKF', 'BLKA', 'FC', 'FD', 'PlusMinus', 'PIR']
data = data[new_order]
data
```

```python
data['X3PT_Per'] = data['X3PT_Per'].str.replace('%', '')
data['X3PT_Per'] = data['X3PT_Per'].str.replace(',', '.')
data['X3PT_Per'] = pd.to_numeric(data['X3PT_Per'])

data['X2PT_Per'] = data['X2PT_Per'].str.replace('%', '')
data['X2PT_Per'] = data['X2PT_Per'].str.replace(',', '.')
data['X2PT_Per'] = pd.to_numeric(data['X2PT_Per'])

data['FT_Per'] = data['FT_Per'].str.replace('%', '')
data['FT_Per'] = data['FT_Per'].str.replace(',', '.')
data['FT_Per'] = pd.to_numeric(data['FT_Per'])

data['height'] = data['height'].str.replace(',', '.')
data['height'] = pd.to_numeric(data['height'])
data[:4]
```

```python
data.to_csv('data_preprocessed.csv', index=False)
```
